```python
import transformers
import torch
```

```python
model_id = "meta-llama/Meta-Llama-3-8B-Instruct"
hf_token = userdata.get('HF_TOKEN')
```

```python
!pip install accelerate
```

```python
pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    token=hf_token,
    device_map="auto",
)
```

```python
messages = [
    {"role": "system", "content": "You are a novel reader who have read many novels! Now you are answering questions about a novel!"},
    {"role": "user", "content": f"""
            Answer the query based on the novel context provided!
            Query : What is backlund?
            Context : {context}
    """},
]
```

```python
pipeline("HeY!")
```

```python
terminators = [
    pipeline.tokenizer.eos_token_id,
    pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
]
```

```python
outputs = pipeline(
    messages,
    max_new_tokens=256,
    eos_token_id=terminators,
    do_sample=True,
    temperature=0.6,
    top_p=0.9,
)
print(outputs[0]["generated_text"][-1])
```
